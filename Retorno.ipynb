# Calculando O Risco de uma carteira

```python
from datetime import datetime
import yfinance as yf

end_date = datetime.now().strftime('%Y-%m-%d')
ativos = ["PSSA3.SA", "AMER3.SA", "KLBN3.SA", "ITSA3.SA", "BEES3.SA", "VGHF11.SA"]

sec_data = yf.download(ativos, start="2019-01-01", progress=False)['Adj Close']
```

```python
sec_data.tail()
```

```python
sec_returns = np.log(sec_data / sec_data.shift(1))
sec_returns
```

```python
sec_returns[["PSSA3.SA", "AMER3.SA", "KLBN3.SA", "ITSA3.SA", "BEES3.SA", "VGHF11.SA"]].mean() * 250
```

```python
weights = np.array([0.11, 0.03, 0.03, 0.11, 0.70, 0.12])
```

Carteira VariÃ¢ncia: A variÃ¢ncia de ativos Ã© um conceito importante no campo de finanÃ§as e investimentos. Ela Ã© uma medida estatÃ­stica que quantifica a dispersÃ£o ou volatilidade dos retornos de um ativo financeiro ao longo do tempo. Quanto maior a variÃ¢ncia, maior a flutuaÃ§Ã£o dos retornos, o que geralmente indica um nÃ­vel de risco mais alto associado ao ativo.

```python
cart_var = np.dot(weights.T, np.dot(sec_returns.cov() * 250, weights))
cart_var
```

Carteira Volatividade : 
A volatilidade de ativos Ã© uma medida da variaÃ§Ã£o dos preÃ§os de um ativo ao longo do tempo. Ela Ã© frequentemente utilizada para avaliar o risco associado a um investimento. Quanto maior a volatilidade de um ativo, maior Ã© a probabilidade de seus preÃ§os flutuarem significativamente em curtos perÃ­odos de tempo.

```python
cart_vol = (np.dot(weights.T, np.dot(sec_returns.cov() * 250, weights))) ** 0.5
cart_vol
```

```python
print(str(round(cart_vol, 5) * 100) + ' %')
```
